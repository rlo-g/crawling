```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 1. req를 통해 멜론차트 정보 요청

head = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36'}

res = req.get("https://www.melon.com/chart/", headers = head)
```

```python
# 2. bs를 통해서 html 데이터로 변형

soup = bs(res.text, "lxml")
soup
```

```python
# 3. 노래 제목 수집
# 클래스 값 중간에 공백 존재 > 클래스가 복수개
song = soup.select("div.ellipsis.rank01 > span > a")      # "cllipsis rank01" 두개의 클래스 사이에 있는 <공백은 .으로 대체>
```

##### 복수 클래스
- 클래스 "" 영역에 공백을 통해 여러개의 클래스가 들어감
- 스타일에는 중복되는 게 많기 때문에 복수개의 클래스를 사용해서 더 용이하게 함


```python
# 4. 가수명 수집

singer = soup.select("div.ellipsis.rank02 > span.checkEllipsis") # 중복, 듀엣 나눠서 출력 방지

# 데이터의 개수가 정해진 경우에는 반드시 데이터 검증이 필요하다.
#   - 데이터의 길이, 중복여부 등..
print(len(singer))
for i in singer:
    print(i.text)
```

```python
# 5. 수집한 데이터를 텍스트 정보만 추출해서 저장

song_list = []
singer_list = []

for i in range(len(song)):
    song_list.append(song[i].text)
    singer_list.append(singer[i].text)
```

```python
# 6. 표 제작

import pandas as pd

dic = {"가수명":singer_list, "노래명":song_list}
chart = pd.DataFrame(dic)
chart
```

```python
# 7. 파일로 제작
#  - 인코딩 작업 : 한글이 깨지지 않게 반드시 해줘야 함
#    1) utf-8
#    2) euc-kr
#    3) utf-8-sig

chart.to_csv("멜론차트.csv", encoding='euc-kr')
```
